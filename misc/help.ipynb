```python
import matplotlib.pyplot as plt 
import seaborn as sns 

df = sns.load_dataset("penguins")
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(1,2, figsize = (12,4))

sns.scatterplot(data = df, x = "bill_length_mm", y = "bill_depth_mm", ax = ax[0], hue = "sex")
sns.scatterplot(data = df, x = "bill_length_mm", y = "flipper_length_mm", ax = ax[1], hue = "sex")

fig.tight_layout()
```
